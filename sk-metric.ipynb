```python
#import pandas as pd
# import math#математические вычисления
import numpy as np#массивы
import matplotlib.pyplot as plt#рисует графики
from matplotlib.colors import ListedColormap#карта цветов
from sklearn import neighbors, datasets #добавляем классификцию ближайших соседей и демонстрационные выборки
from sklearn.model_selection import LeaveOneOut # дробит выборку на все возможные варианты без одного тестового элемента
# from sklearn.metrics import accuracy_score # вычисляет точность классификации
```

```python
# загрузка данных "Ирисы фишера"
iris = datasets.load_iris()
y=iris.target # номера классов
X=iris.data[:,[0,3]] # 2 признака
```

```python
h = .01  # шаг сетки
# создание цветовых карт
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

<p>Для начала попробуем построить самые эффективные <i>(минимум ошибок по результатам Loo)</i> классификации в metric : default ‘minkowski’ <i>(Метрика по умолчанию, а так как по умолчанию p=2, это декартово расстояние)</i>  для:</p>
<ol>
<li>weights = 'uniform' это когда веса распределяются одинаково для всех элементов</li>
<li>weights = 'distance' это когда веса обратнопропорциональны расстояниям от классифицируемого элемента</li>
</ol>



```python
Loo=LeaveOneOut()
for weights in ['uniform', 'distance']:
    Q=[] # список ошибок соответстующий числам соседей от 1 до на 1 меньше чем элементов в выборке
    for n in range(len(y)-1): #Для Loo максимальное число соседей на 1 меньше главной выборки т.к. 1 элемент извлекается
        qq=0 # число ошибок в текущем скользящем контроле
        clf = neighbors.KNeighborsClassifier(n+1, weights=weights) #создаем экземпляр классификатора
        for train_index, test_index in Loo.split(X): #разбиваем индексы на тестовые и обучающие
            clf.fit(X[train_index], y[train_index]) #подгоняем данные
            qq+= (y[test_index]!=clf.predict(X[test_index])).sum() #увеличиваем число ошибок если тест не равен предсказанию
        Q.append(qq) #Добавляем частоту ошибок в список частот ошибок
    minQ=min(Q) # ищем минимальное количство ошибок
    n_neighbors=Q.index(minQ)+1 # ищем минимальное число соседей при минимуме ошибок
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) #создаем экземпляр классификатора
    clf.fit(X, y) #подгоняем данные
    #Для графического представления, найдем максимальные и минимальные значения признаков и увеличим(уменьшим) их на 1
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(), X[:, 1].max()
    # получим сетку
    xx, yy = np.meshgrid(np.arange(x_min - 1, x_max + 1, h),
                         np.arange(y_min - 1, y_max + 1, h))
    # метод ravel "распутывает" матрицу в массив приставляя к первой строке вторую потом третью и т.д.
    # метод c_ принимает два массива и превращает их в матрицу, где эти массивы столбцы
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # классифицыруем сетку

    # аргумент shape выдает форму матрицы (число строк, число столбцов)
    # метод reshape принимает форму и переделывает в соответствии с ней
    Z = Z.reshape(xx.shape)
    
    fig, axs = plt.subplots(nrows=1, ncols=2, sharex=False) # создаем одну фигуру с двумя графиками по горизонтали
    
    fig.set_size_inches(29.7 / 2.54 , 21. / 2.54) # установить размер изображения формат А4, альбомная ориентация
    
    axs[0].plot(range(1,len(Q)+1),Q) # подаем в фигуру массивы соседей и соответствующих ошибок
    axs[0].set_title("Q(n), Qmin = %i" %minQ) # заголовок графика
    axs[0].set_xlabel("число соседей") # подпись горизонтальной оси
    axs[0].set_ylabel("число ошибок") # подпись вертикальной оси
    axs[0].scatter(n_neighbors, minQ, c='none', edgecolor='r', linewidths=2, s=25) # обводим кружком первую точку минимума
    # подписываем точку чуть ниже
    axs[0].text(n_neighbors, minQ-(max(Q)-minQ)/100, "min", fontsize=12, horizontalalignment='center', verticalalignment='top')
    
    axs[1].pcolormesh(xx, yy, Z, cmap=cmap_light) #заливаем фон разными цветами, на основе сетки, классификации и цветовой карты

    # раставляем элементы выборки кружками радиуса 35 с черным ободком
    axs[1].scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=35) # наносим известные элементы на график
    filtr= y!=clf.predict(X) # получим логический фильтр неправильно классифицированных элементов
    XErr = X[filtr] # массив неправильноклассифицированных элементов
    yErr = y[filtr] # соответствующие им правильные ответы
#     axs[1].scatter(XErr[:, 0], XErr[:, 1], c='none', linewidths=3, edgecolor='y', s=60)
    # зачеркиваем неправильтно классифицированные элементы красными "иксами"
    for Er in XErr:
        axs[1].text(Er[0], Er[1], "X", color="r", fontsize=16, horizontalalignment='center', verticalalignment='center')
    # установка ограничений на оси, немного обрежем по краям график, чтобы увеличиласть область с известными ответами
    axs[1].set_xlim(x_min-0.5, x_max+0.5)
    axs[1].set_ylim(y_min-0.5, y_max+0.5)
    # добавление заголовка
    axs[1].set_title("knn k = %i, weights = '%s')"
              % (n_neighbors, weights))
    plt.show()
```

<p>Метод Парзеновского окна Должен для каждого ядра:</p>
<ol>
<li>E - Епанечникова</li>
<li>Q - Квартическое</li>
<li>T - Треугольное</li>
<li>G - Гауссовское</li>
<li>П - Прямоугольное</li>
</ol>
<p>с помощью скользящего контроля найти оптимальную h - ширину окна.</p>
<p>Суть ширины окна, в том, что когда ядро Финитное, при подстановки в него отношения расстояния к ширине окна будем получать нулевые веса всем обьектам на расстоянии больше h.</p>
<p>Подобный эффект можно получить применив <i>RadiusNeighborsClassifier</i>, но мы не ищем легких путей и будем добавлять ядра в <i>KNeighborsClassifier</i>.</p>

```python
# Создадим функцию которая приняв перечисляемый(поддерживает индексы [индекс] и слайсы [начало:конец:шаг]) объект
# вернет такой же объект но заполненный не расстояниями, а весами высчитанными через ядра,
# которые будут указываться отдельным пармаетром
from math import exp,sqrt,pi
from copy import deepcopy

def weights_using_kernel(distances, R=1, kernel='Epanechnikov'):
    """Выдает элемент аналогичный принятому distances где на месте соответсвующих стоят веса, посчитанные
    как K(p/h), где h - радиус окна, p - расстояние являющееся элементом distances и K - ядро.
    поддерживает ядро Епанечникова, Квартическое, Треугольное, Гауссовское, Прямоугольное, соответствующие ключи:
    'Epanechnikov', 'Quartic', 'Triangular', 'Gaussian', 'Uniform' """
    if kernel == 'Epanechnikov':
        K=(lambda el: (3. * (1. - el**2) / 4.) if abs(el)<1. else 0.0)
    elif kernel == 'Quartic':
        K=lambda el: 15. / (0.25 - el**2 / 4.)**2 if abs(el)<1. else 0.0
    elif kernel == 'Triangular':
        K=lambda el: (1. - abs(el)) if abs(el)<1. else 0.0
    elif kernel == 'Gaussian':
        K=lambda el: exp(-1. * el**2 / 2.) / sqrt(2. * pi)
    elif kernel == 'Uniform':
        K=lambda el: 0.5 if abs(el)<1. else 0.0
    else:
        raise AssertionError("Такое ядро не предусмотрено! ", kernel)
    return np.vectorize(K)(distances/R)
```

```python
R_max = (((x_max-x_min)**2 + (y_max - y_min)**2)**0.5) / 2. #верхняя оценка максимального радиуса
n_neighbors=len(X)-1
lst_R = tuple([0.05*float(x) for x in range(1,int(R_max/0.05)+1)])
```

```python
for kernel in ['Epanechnikov', 'Quartic', 'Triangular', 'Gaussian', 'Uniform']:
    Q=[] # список ошибок соответстующий радиусам
    
    for radius in lst_R:
        qq=0 # число ошибок в текущем скользящем контроле
        #создаем экземпляр классификатора
        fun_weight=lambda dist: weights_using_kernel(dist, radius, kernel)
        clf = neighbors.KNeighborsClassifier(n_neighbors, fun_weight)
        for train_index, test_index in Loo.split(X): #разбиваем индексы на тестовые и обучающие
            clf.fit(X[train_index], y[train_index]) #подгоняем данные
            qq+= (y[test_index]!=clf.predict(X[test_index])).sum() #увеличиваем число ошибок если тест не равен предсказанию
        Q.append(qq) #Добавляем частоту ошибок в список частот ошибок
    imin=0
    minQ=Q[0]
    for i in range(len(Q)):
        if Q[i]<minQ:
            minQ=Q[i]
            imin=i
    opt_R=lst_R[imin]
    fun_weight=lambda dist: weights_using_kernel(dist, opt_R, kernel)
    clf = neighbors.KNeighborsClassifier(n_neighbors, fun_weight) #создаем экземпляр классификатора
    clf.fit(X, y) #подгоняем данные
    #Для графического представления, найдем максимальные и минимальные значения признаков и увеличим(уменьшим) их на 1
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    # получим сетку
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    # метод ravel "распутывает" матрицу в массив приставляя к первой строке вторую потом третью и т.д.
    # метод c_ принимает два массива и превращает их в матрицу, где эти массивы столбцы
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # классифицыруем сетку

    # аргумент shape выдает форму матрицы (число строк, число столбцов)
    # метод reshape принимает форму и переделывает в соответствии с ней
    Z = Z.reshape(xx.shape)
    
    fig, axs = plt.subplots(nrows=1, ncols=2, sharex=False) # создаем одну фигуру с двумя графиками по горизонтали
    
    fig.set_size_inches(29.7 / 2.54 , 21. / 2.54) # установить размер изображения формат А4, альбомная ориентация
    
    axs[0].plot(lst_R,Q) # подаем в фигуру массивы окон и соответствующих ошибок
    axs[0].set_title("Q(h), Qmin = %i, h = %.2f" %(minQ,opt_R)) # заголовок графика
    axs[0].set_xlabel("ширина окна") # подпись горизонтальной оси
    axs[0].set_ylabel("число ошибок") # подпись вертикальной оси
    axs[0].scatter(opt_R, minQ, c='none', edgecolor='r', linewidths=2, s=25) # обводим кружком первую точку минимума
    # подписываем точку чуть ниже
    axs[0].text(opt_R, minQ-(max(Q)-minQ)/100, "min", fontsize=12, horizontalalignment='center', verticalalignment='top')
    
    axs[1].pcolormesh(xx, yy, Z, cmap=cmap_light) #заливаем фон разными цветами, на основе сетки, классификации и цветовой карты

    # раставляем элементы выборки кружками радиуса 35 с черным ободком
    axs[1].scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=35) # наносим известные элементы на график
    filtr= y!=clf.predict(X) # получим логический фильтр неправильно классифицированных элементов
    XErr = X[filtr] # массив неправильноклассифицированных элементов
    yErr = y[filtr] # соответствующие им правильные ответы
#     axs[1].scatter(XErr[:, 0], XErr[:, 1], c='none', linewidths=3, edgecolor='y', s=60)
    # зачеркиваем неправильтно классифицированные элементы красными "иксами"
    for Er in XErr:
        axs[1].text(Er[0], Er[1], "X", color="r", fontsize=16, horizontalalignment='center', verticalalignment='center')
    # установка ограничений на оси, немного обрежем по краям график, чтобы увеличиласть область с известными ответами
#     axs[1].set_xlim(xx.min()+0.5, xx.max()-0.5)
#     axs[1].set_ylim(yy.min()+0.5, yy.max()-0.5)
    # добавление заголовка
    axs[1].set_title("Parzen window wis kernel = '%s')"
              % (kernel))
    plt.show()
```

```python
# Найдем "наложившиеся" элементы выборки
dicX=dict()
for ind in range(len(X)):
    tX=tuple(X[ind])
    if tX not in dicX:
        dicX[tX]=[ind]
    else:
        dicX[tX].append(ind)
suspect = [x for x in dicX if len(dicX[x])>2]

dicY=dict()
for tup in suspect:
    dicY[tup]=[y[ind] for ind in dicX[tup]]
for tu in dicY:
    print(tu,end=' ');print(dicY[tu],end=' ');print(dicX[tu])
```

Как видим 134 элемент относится к классу 2"синие", а 91 и 63 к классу 1(зелёные), все классификаторы определяют эти координаты как 1(зелёные) класс, следовательно вне зависимости от применяемых методов один элемент будет ошибочно классифицирован.

По результатам Loo самыми точными оказались knn при k = 16, метод парзеновского окна с треугольным и Епанечникова ядрами, они классифицировали всего с 5ю ошибками.

В методе Парзеновского окна бросилось в глаза, что большая часть плоскости классифицируется, как 0(красный) класс, это связано с тем что при прочих равных голосах, побеждает самый первый класс, а финитные ядра делают все голоса нулевыми если в окно не попал ни один элемент, правда этого недостатка лишено Гаусовское ядро потому что оно не "отрубает" от голосования элементы не попавшие в радиус

Можно сделать радиус переменным и сделать таким, чтобы в окно всегда попадали k или более(если невозможно ровно к) элемнтов,
но при таком подходе метод парзеновского окна станет очень сильно похож на knn сглаженый ядром.
